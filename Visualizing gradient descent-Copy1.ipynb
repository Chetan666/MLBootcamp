```python
import numpy as np

# Define the neural network class
class NeuralNetwork:
    def __init__(self):
        # Initialize the weights and biases for a simple single-layer network
        self.input_size = 1
        self.output_size = 1
        self.learning_rate = 0.01

        self.W2 = np.random.randn(self.input_size, self.output_size)
        self.b2 = np.zeros((1, self.output_size))

    def forward(self, x):
        # Forward pass
        self.z2 = np.dot( x,self.W2) + self.b2
        self.output = self.z2 #identity activation function
        outputarr.append(self.output[0])
        return self.output

    def backward(self, x, y):
        # Backpropagation
        m = x.shape[0]
        #print("m=",m)
        error = self.output - y

        dW2=-2*np.dot(self.output.T, error) / m
        db2 = np.sum(error, axis=0) / m



        # Update weights and biases
        self.W2 -= self.learning_rate * dW2
        self.b2 -= self.learning_rate * db2

    def train(self, x, y, epochs):
        for epoch in range(epochs):
            # Forward and backward pass
            output = self.forward(x)
            #outputarr.append(output)
            self.backward(x, y)

            # Calculate and print the mean squared error
            mse = np.mean((output - y) ** 2)
            msearr.append(mse)
            print(f"Epoch {epoch + 1}/{epochs}, MSE: {mse:.4f}")

# Generate some example data
msearr=[]
outputarr=[]
x=np.array([0.2]) #0.2 is input N1_1 for the NN we discussed
y=1 #Predicted output

# Create and train the neural network
model = NeuralNetwork()
epochs=500
model.train(x, y, epochs) #increase epochs for more accuracy. 
```

```python
import matplotlib.pyplot as plt

# Plot MSE w.r.t Error
#plt.figure(figsize=(8, 6))
fig=plt.figure()
plt.title('Visualizing Gradient Descent-1')
plt.xlabel('Epoch')
plt.ylabel('MSE')
indices = range(len(msearr))
plt.plot(indices, msearr, marker='o', linestyle='-')
plt.savefig('MSE.png')
```
