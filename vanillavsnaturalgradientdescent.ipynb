```python
import numpy as np
import matplotlib.pyplot as plt

def quadratic_objective(theta, A):
    return 0.5 * np.dot(np.dot(theta, A), theta)

def gradient_quadratic_objective(theta, A):
    return np.dot(A, theta)

def natural_gradient_descent(initial_theta, A, learning_rate, num_iterations):
    theta_values = [initial_theta]

    for _ in range(num_iterations):
        gradient = gradient_quadratic_objective(theta_values[-1], A)
        fisher_matrix_inverse = np.linalg.inv(A)
        natural_gradient = np.dot(fisher_matrix_inverse, gradient)
        updated_theta = theta_values[-1] - learning_rate * natural_gradient
        theta_values.append(updated_theta)

    return np.array(theta_values)

def vanilla_gradient_descent(initial_theta, A, learning_rate, num_iterations):
    theta_values = [initial_theta]

    for _ in range(num_iterations):
        gradient = gradient_quadratic_objective(theta_values[-1], A)
        updated_theta = theta_values[-1] - learning_rate * gradient
        theta_values.append(updated_theta)

    return np.array(theta_values)

# Parameters
A = np.array([[2.0, 0.5], [0.5, 1.5]])  # Positive-definite matrix
initial_theta = np.array([2.0, 3.0])
learning_rate = 0.2
num_iterations = 50

# Run natural gradient descent
trajectory_natural_gradient = natural_gradient_descent(initial_theta, A, learning_rate, num_iterations)

# Run vanilla gradient descent
trajectory_vanilla_gradient = vanilla_gradient_descent(initial_theta, A, learning_rate, num_iterations)

# Plot the results
plt.plot(trajectory_natural_gradient[:, 0], trajectory_natural_gradient[:, 1], label='Natural Gradient Descent', marker='o')
plt.plot(trajectory_vanilla_gradient[:, 0], trajectory_vanilla_gradient[:, 1], label='Vanilla Gradient Descent', marker='x')
plt.title('Comparison of Natural Gradient Descent and Vanilla Gradient Descent')
plt.xlabel('Theta 1')
plt.ylabel('Theta 2')
plt.legend()
plt.show()
```
